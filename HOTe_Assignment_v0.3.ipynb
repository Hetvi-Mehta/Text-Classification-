```python
path =r'D:\\INSOFE\\CSE 7124c Text Mining\\HOT\\Combined'

def ffilter(files, pattern, root, i):
    data=[]
    for fname in fnmatch.filter(files, pattern):
        data_file = open(os.path.join(root, fname), "r")
        text = data_file.read();
        data.append(text) 
    doc = pd.DataFrame(data)
    doc[1] = i
    return(doc);
```

```python
for root, dirnames, filenames in os.walk(path):
    d1 = ffilter(filenames, '*_1', root, 1) 
    d2 = ffilter(filenames, '*_2', root, 2)
    d3 = ffilter(filenames, '*_3', root, 3)
    d4 = ffilter(filenames, '*_4', root, 4)
    d5 = ffilter(filenames, '*_5', root, 5)
    d6 = ffilter(filenames, '*_6', root, 6)
df = [d1, d2, d3, d4, d5, d6]
data_set = pd.concat(df, ignore_index=True)
data_set.columns = ['Text', 'Document']
```

```python
text_data = df['Text']
text_list = text_data.tolist()   
```

```python
for i in range(len(text_list)):
    text_list[i] = re.sub(r'\s+', ' ', text_list[i]) #Removing more than one white spaces in the sentence
    text_list[i] = re.sub('[\d]', '', text_list[i]) #Removing numbers
    text_list[i] = re.sub("\n", '', text_list[i])
```

```python
#look for 2 or more repetitions of character
recc = re.compile(r"(.)\1{1,}", re.DOTALL)
```

```python
for i in range(len(text_list)):
    text_list[i]= recc.sub(r"\1\1", text_list[i])
    text_list[i]= text_list[i].lower()
    text_list[i] = text_list[i].strip('\'"?,.')
    text_list[i] =re.sub(r'[^a-zA-Z]',r' ', text_list[i])
print(text_list[0:20])
```

```python
all_words = []
sub_list = []
for item in text_list:
    tokenized = nltk.word_tokenize(item)
    tagged = nltk.pos_tag(tokenized)
    allowed = ['JJ']
    for w in tagged:
        if w[1] in allowed:
            all_words.append(w[0])
    all_words.append(sub_list)
print(all_words)
```

```python
##Metrics: Multiclassification Metric 
from sklearn.metrics import confusion_matrix
from sklearn.metrics import recall_score, precision_score, accuracy_score

def MetricCalculation(y_train,pred_train,y_test,pred_test,ModelMethodName):

    prediction=confusion_matrix(y_train, pred_train) #get confusion matrix for train predictions
    print("Train_Confusion Matrix:\n",prediction,"\n")

    acc = accuracy_score(y_train, pred_train)
    rec = recall_score(y_train, pred_train,average=None)
    prec = precision_score(y_train, pred_train,average=None)

    print("Results of",ModelMethodName," on train data:","\nAccuracy:",acc, "\nRecall:",rec, "\nPrecision:",prec , "\n")
    
    prediction = confusion_matrix(y_test,pred_test) ##get confusion matrix for test predictions
    print("Test_Confusion matrix:\n", prediction,"\n")

    acc = accuracy_score(y_test,pred_test)
    rec = recall_score(y_test,pred_test,average=None)
    prec=precision_score(y_test,pred_test,average=None)
    print("Results of ",ModelMethodName," on test data:","\nAcc: ",acc,"\nRec:", rec, "\nPrec:", prec,"\n")
```

# Naive Bayes 

```python
from sklearn.naive_bayes import MultinomialNB

def NaiveBayes(X_train,X_test,y_train,y_test):
    nb_clf = MultinomialNB().fit(X_train, y_train) #nb_clf is the classifier
    pred_train = nb_clf.predict(X_train) #get predictions on train data
    pred_test = nb_clf.predict(X_test) #predict on test data    
    print("Naive Bayes")
    MetricCalculation(y_train,pred_train,y_test,pred_test,"Naive Bayes' ")
```

# Logistic Regression

```python
from sklearn import linear_model
def LogisticRegression(X_train,X_test,y_train,y_test):
    logreg_train = linear_model.LogisticRegression()
    lr_clf = logreg_train.fit(X_train, y_train) #fit logreg model on train data
    pred_lr_train=lr_clf.predict(X_train) # predict on train data
    pred_lr_test=lr_clf.predict(X_test) # predict on test data
    print("Logistic Regression")
    MetricCalculation(y_train,pred_lr_train,y_test,pred_lr_test,"Logistic Regression ")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
```

# Modelling

```python
####Count Vectorizer
count=CountVectorizer(strip_accents='unicode',ngram_range=(1,1),stop_words='english')
tfreq=count.fit_transform(X_train.Text)
Dense_mat=tfreq.todense()
X_train_df=pd.DataFrame(Dense_mat,columns=count.get_feature_names())

print(type(X_train_df))
print(X_train_df.shape)

# Get the tfidf matrix for test documents
X_test_df = count.transform(X_test.Text)
print(type(X_test_df))
print(X_test_df.shape)

#print(X_train_df.head())
#print(X_train_df.tail())

NaiveBayes(X_train_df,X_test_df,y_train,y_test)
LogisticRegression(X_train_df,X_test_df,y_train,y_test)
```

```python
###TFIDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(ngram_range=(1,1),stop_words='english',vocabulary=all_words)
#print(tfidf_vectorizer.vocabulary_)
#print(tfidf_vectorizer.idf_)

tfidf = tfidf_vectorizer.fit_transform(X_train.Text)
Dense_mat = tfidf.todense()
X_train_tfidf=pd.DataFrame(Dense_mat,columns=tfidf_vectorizer.get_feature_names())

print(type(X_train_tfidf))
print(X_train_tfidf.shape)

# Get the tfidf matrix for test documents
X_test_tfidf = tfidf_vectorizer.transform(X_test.Text)
print(type(X_test_tfidf))
print(X_test_tfidf.shape)

#print(X_train_tfidf.head())
#print(X_train_tfidf.tail())

NaiveBayes(X_train_tfidf,X_test_tfidf,y_train,y_test)
LogisticRegression(X_train_tfidf,X_test_tfidf,y_train,y_test)
```
